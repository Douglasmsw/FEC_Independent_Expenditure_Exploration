```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter('ignore', FutureWarning)

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
full_data = pd.read_csv('https://cg-519a459a-0ea3-42c2-b7bc-fa1143481f74.s3-us-gov-west-1.amazonaws.com/bulk-downloads/2022/independent_expenditure_2022.csv')
print(full_data.shape)
full_data=full_data.fillna(0)
full_data.head(50)
```

```python
select_data = pd.DataFrame()
select_data['spe_nam'] = full_data['spe_nam'].str.upper()
select_data['cand_pty_aff'] = full_data['cand_pty_aff']
select_data['sup_opp'] = full_data['sup_opp']
select_data['exp_amo'] = full_data['exp_amo']
select_data['cand_name'] = full_data['cand_name']
select_data['can_office_state'] = full_data['can_office_state']
select_data['can_office'] = full_data['can_office']
select_data['use'] = full_data['pur']
select_data['cand_name'] = select_data['cand_name'].str.upper()

select_data.head(5)
```

```python
pty_affil_clean = select_data.copy()

for i in range (len(pty_affil_clean)):
    if pty_affil_clean['cand_pty_aff'][i] == 0:
        pty_affil_clean = pty_affil_clean.drop([i])

pty_affil_clean.reset_index(drop=True, inplace=True)
print(pty_affil_clean.shape)
pty_affil_clean.head(5)
```

```python
pty_clean = pty_affil_clean.copy()

for k in range (len(pty_clean)):
    if str(pty_clean['cand_pty_aff'][k]) == 'REPUBLICAN PARTY':
        if str(pty_clean['sup_opp'][k]) == 'S':
            pty_clean = pty_clean.drop([k])
    elif str(pty_clean['cand_pty_aff'][k]) == 'DEMOCRATIC PARTY':
        if str(pty_clean['sup_opp'][k]) == 'O':
            pty_clean = pty_clean.drop([k])

pty_clean.reset_index(drop=True, inplace=True)    
print(pty_clean.shape)
pty_clean.head(50)
```

```python
def state_sweeper(state):
    state_clean = pty_clean.copy()
    for j in range (len(state_clean)):
        if str(state_clean['can_office_state'][j]) != state:
            state_clean = state_clean.drop([j])
    state_clean.reset_index(drop=True, inplace=True)    
    spender_list = state_clean.groupby(["spe_nam", "sup_opp", "cand_name"]).sum()[['exp_amo']]
    spender_list.reset_index(inplace=True)
    spender_list = spender_list.sort_values(by=['exp_amo'], ascending=False)
    spender_list.reset_index(drop=True, inplace=True)
    spender_list.index = spender_list.index + 1
    print(state)
    return spender_list.head(10)
        

state = 'NC'
state_sweeper(state)
```

```python
def state_sweeper(state):
    state_clean = pty_clean.copy()
    for j in range (len(state_clean)):
        if str(state_clean['can_office_state'][j]) != state:
            state_clean = state_clean.drop([j])
    state_clean.reset_index(drop=True, inplace=True)    
    spender_list = state_clean.groupby(["cand_name", "sup_opp", "cand_pty_aff"]).sum()[['exp_amo']]
    spender_list.reset_index(inplace=True)
    spender_list = spender_list.sort_values(by=['exp_amo'], ascending=False)
    spender_list.reset_index(drop=True, inplace=True)
    spender_list.index = spender_list.index + 1
    print(state)
    return spender_list
        

state = 'WI'
state_sweeper(state)

```

```python
def state_cost(state_lst):
    state_expo = []
    for j in range (len(state_lst)):
        state_clean = pty_clean.copy()
        for k in range(len(state_clean)):
            if str(state_clean['can_office_state'][k]) != str(state_lst[j]):
                state_clean = state_clean.drop([k])
        state_expo.append(sum(state_clean['exp_amo']))
    state_expo_rank = pd.DataFrame()
    state_expo_rank['state'] = state_lst
    state_expo_rank['total_expo'] = state_expo
    state_expo_rank = state_expo_rank.sort_values(by=['total_expo'], ascending=False)
    state_expo_rank.reset_index(drop=True, inplace=True)
    state_expo_rank.index = state_expo_rank.index + 1
    return state_expo_rank

state_lst = ['AZ', 'CO', 'FL', 'GA', 'IA', 'ME', 'MI', 'MN', 'NC', 'NV', 'OH', 'PA', 'VA', 'WI']
state_cost(state_lst)
```
